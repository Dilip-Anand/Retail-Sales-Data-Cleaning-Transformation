```
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.util import *
```

#Loading the RAW Data

```
df = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/tables/retail_sales_dataset.csv")
```

```
display(df)
```

#Handle missing values (impute, drop, or fill).

```
df.count()
```

```
# Drop rows missing critical fields
df = df.dropna(subset=['Transaction ID', 'Date', 'Customer ID', 'Product Category', 'Quantity', 'Price per Unit'])
```

```
# Fill missing Gender with 'Unknown'
df = df.fillna({'Gender': 'Unknown'})

# Fill missing Age with median
df = df.withColumn("Age",col("Age").cast("int"))
median_age = df.approxQuantile('Age', [0.5], 0.01)[0]
df = df.fillna({'Age': median_age})
```

#Remove duplicates

```
df = df.dropDuplicates()
```

```
df.count()
```

#Convert the columns from string to other date type.

```
df = df.withColumn("Date",col("Date").cast("date"))
```

```
df = df.withColumn("Transaction ID", col("Transaction ID").cast("int")) \
       .withColumn("Quantity", col("Quantity").cast("int"))\
       .withColumn("Price per Unit", col("Price per Unit").cast("int"))\
        .withColumn("Total Amount", col("Total Amount").cast("int"))\
       .orderBy("Transaction ID")
```

```
display(df)
```

#Create final Data to Calculate monthly sales, revenue

```
df = df.withColumn("Month",month(col("Date")))
```

```
df_Monthly = df.groupBy("Month").agg(sum("Total Amount").alias("Total_Monthly_Revenue"),sum("Quantity").alias("Monthly_Quantity_Sold"),count("Transaction ID").alias("Monthly_Sales")).orderBy("Month")
```

```
display(df_Monthly)
```

```
pdf = df_Monthly.toPandas()
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(pdf["Month"], pdf["Total_Monthly_Revenue"], color="green")
plt.xlabel("Month")
plt.ylabel("Total Monthly Revenue in (Rs)")
plt.title("Monthly Revenue Overview")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(10, 5))
plt.bar(pdf["Month"], pdf["Monthly_Sales"], color="blue")
plt.xlabel("Month")
plt.ylabel("Total Monthly Sales")
plt.title("Monthly Sales Overview")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
import plotly.express as px

fig = px.scatter_3d(
    pdf,
    x="Month",
    y="Monthly_Sales",
    z="Total_Monthly_Revenue",
    color="Monthly_Quantity_Sold",  # Optional: adds color dimension
    title="3D Monthly Sales Analysis"
)
fig.show()
```
